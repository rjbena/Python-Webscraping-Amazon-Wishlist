Short python webscraping project to log multiple amazon products' price changes to a csv file.

```python
#Testing to get data from a single url.

import requests 
from bs4 import BeautifulSoup
import time

URL = 'https://www.amazon.com/Data-Analytics-Engineering-Analyst-T-Shirt/dp/B08WD95TMV/ref=sr_1_5?crid=3VU0BE333RG4E&keywords=data+analytics+shirt&qid=1652205515&sprefix=data+analytics+sh%2Caps%2C145&sr=8-5'
headers ={"User-Agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36'}

#print(response.content)

while(1):
    try:
        response = requests.get(URL, headers = headers)
        soup = BeautifulSoup(response.content, 'html.parser')

        title = soup.find('h1').get_text()
        print(title.strip())

        price =float(soup.find('span', class_='a-offscreen').get_text()[1:])
        print(price)

        rating = soup.find('span', class_='a-icon-alt').get_text()
        rating= float((rating)[:3])
        print(rating)
    except AttributeError:
        print('retrying...')
    else:
        break
```

```python
# Creating file to write data into.

import csv

header = ["Product Name", "Date", "Price", "Rating"]

with open("amazon_wishlist_dataset.csv", "w", encoding="UTF-8",newline="") as f:
    writer = csv.writer(f)
    writer.writerow(header)
```

```python
# Function to write data into csv file.

import csv 
import requests 
import datetime 
from bs4 import BeautifulSoup


def check_wishlist():
    URL = 'https://www.amazon.com/Data-Analytics-Engineering-Analyst-T-Shirt/dp/B08WD95TMV/ref=sr_1_5?crid=3VU0BE333RG4E&keywords=data+analytics+shirt&qid=1652205515&sprefix=data+analytics+sh%2Caps%2C145&sr=8-5'
    headers ={"User-Agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36'}
    while(1):
        try:
            response = requests.get(URL, headers = headers)
            soup = BeautifulSoup(response.content, 'html.parser')
            
            title = soup.find('h1').get_text().strip()
            

            price =float(soup.find('span', class_='a-offscreen').get_text()[1:])
            print(price)

            rating = soup.find('span', class_='a-icon-alt').get_text()
            rating= float((rating)[:3])
            print(rating)
            
            today = datetime.date.today()
            print(today)
            data =[title,today,price,rating]
            
            with open('amazon_wishlist_dataset.csv','a', encoding='utf-8', newline='') as f:
                writer = csv.writer(f)
                writer.writerow(data)
        except AttributeError:
            print('retrying...')
        else:
            break
```
